```python
import os
import json

# --- Configuration ---
# Set this to the path where your .json files are located.
# If the script is in the SAME folder as your json files, you can just use "."
ANNOTATION_FOLDER = "." 
# --- End of Configuration ---

print(f"Scanning for .json files in: {os.path.abspath(ANNOTATION_FOLDER)}")

# Loop through every file in the specified folder
for filename in os.listdir(ANNOTATION_FOLDER):
    # Check if the file is a JSON annotation file
    if filename.endswith(".json"):
        json_filepath = os.path.join(ANNOTATION_FOLDER, filename)
        
        # A flag to check if we made any changes
        was_modified = False

        # Open and read the file
        with open(json_filepath, 'r') as f:
            data = json.load(f)

        # Loop through all the labeled shapes (polygons) in the file
        for shape in data['shapes']:
            # Check if the label starts with 'dog'
            if shape['label'].startswith('dog'):
                # If it does, change it to just 'dog'
                shape['label'] = 'dog'
                was_modified = True
        
        # If we changed any label in this file, save the changes
        if was_modified:
            with open(json_filepath, 'w') as f:
                json.dump(data, f, indent=2)
            print(f"Updated labels in: {filename}")

print("\n✅ Done! All relevant labels have been changed to 'dog'.")
```
