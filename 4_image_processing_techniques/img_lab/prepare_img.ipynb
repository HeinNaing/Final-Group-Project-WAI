```python
import random
import matplotlib.pyplot as plt

test_img_path = r"C:\Users\Lenovo\Documents\GitHub\Practical-ML-by-mysterious-eggs\6_deep_learning\CNN\cats_and_dogs_filtered\test\dogs"
os.listdir(test_img_path)
```

```python
# selected_img = random.choice(os.listdir(test_img_path))
for i in range(2460, 2474):   # 11 to 50 inclusive
    selected_img = f"dog.{i}.jpg"
    print("selected_img :", selected_img)

    selected_img_full_path = test_img_path + "/" + selected_img
    print("selected_img_full_path :", selected_img_full_path)

    read_img = cv2.imread(selected_img_full_path)
    if read_img is None:
        print("Image not found :", selected_img)
        continue  # skip to next

    resize_img = cv2.resize(read_img, (300, 300))
    plt.imshow(cv2.cvtColor(resize_img, cv2.COLOR_BGR2RGB))
    plt.axis("off")
    plt.title(selected_img)
    plt.show()
    
    # Remove .jpg and save as .png
    base_name = os.path.splitext(selected_img)[0]  # "cat.11"
    write_name = f"./data_annotated/test/dogs/{base_name}.png"
    cv2.imwrite(write_name, resize_img)
    print("Saved:", write_name)
```
